#### SDE in FX

It is commonly assumed that the foreign exchange rates fluctuates
as a diffusion process that can be expressed in the form of a Stochastic
Differential Equation (SDE). 

The local volatility model assumes that the coefficients of the SDE
are deterministic functions of the underlying, the exchange rate denoted
$X$. 

$$
dX/X=\left(r_{d}-r_{f}\right)dt+\sigma\left(X,\,t\right)dW_{t}
$$

where $r_{d}$ denotes the domestic interest rate and $r_{f}$ the
foreign interest rate. Note that in dynamic equilibrium, one would
expect the exchange rate to have mean $X_{0}e^{\int\left(r_{d}-r_{f}\right)dt}$
so that there does not exist a statistical arbitrage opportunity.
Nevertheless, it is not clear what deterministic function we should
use for $\sigma$. In fact, one would expect different exchange rate
pairs to have a different function $\sigma$. A common approach in
industry is to find the $\sigma$ `implied` by the European call prices
in the market. 

Necessary and sufficient conditions for existence of such a function
can be found quite easily and intuitively. As long as $K_{2}>K_{1}\implies C(K_{2})\geq C(K_{1})$
then there exists a local volatility function that correctly prices
all European vanilla options. Nevertheless, a local volatility model
typically fails to price correctly path-dependent options such as
barrier or Asian options. One may opt for a stochastic volatility
model as well, but according to Claudia it also does not work very
well. In fact the market prices for path-dependent options (e.g. Upper
One Touch) predominantly sit somewhere between those predicted by
local and stochastic volatility. {[}1{]} Therefore it is natural to
use a model that is a combination of the two, named the Local Stochastic
Volatility (LSV) model. We now present two variations of the LSV model:

Local Stochastic Volatility Heston:

$$
\begin{eqnarray*}
dX/X & = & \mu dt+A(S,t)\sigma dW_{t}\\
d\left(\sigma^{2}\right) & = & \lambda\left(\psi-\sigma^{2}\right)dt+\gamma\sigma dZ_{t}\\
\left\langle dW_{t},\,dZ_{t}\right\rangle  & = & \rho dt
\end{eqnarray*}
$$

Local Stochastic Volatility Hybrid:

$$
\begin{eqnarray*}
dX_{t}/X_{t} & = & \mu dt+A(S,t)\sigma dW_{t}\\
d\sigma_{t} & = & \lambda\left(\sigma_{\infty}-\sigma_{t}\right)dt+\gamma\sigma_{t}dZ_{t}\\
\left\langle dW_{t},\,dZ_{t}\right\rangle  & = & \rho dt
\end{eqnarray*}
$$

Since both models defy an analytic solution, they must be tackled
by solving the Black-Scholes PDE using finite difference. In both
case, the Black Scholes PDE is 
$$
\frac{\partial V}{\partial t}+rV\frac{\partial V}{\partial x}+\left(\sigma A\right)^{2}\frac{\partial^{2}V}{\partial x^{2}}=rV
$$

the problem posed by time-interpolation of the local volatility functions.

Note that in these two models, the Dupire implied volatility, $\sigma_{D}$,
satisfies $\sigma_{D}(t,\,K)=\sigma_{t}A(t,\,K)$. One can prove that
$$
\mathbb{E}\left[\sigma_{T}^{2}A(T,\,S_{T})^{2}\vert S_{T}=K\right]=\sigma_{D}(T,\,K)
$$


thus
$$
A(T,\,K)^{2}=\frac{\sigma_{D}(T,K)^{2}}{\mathbb{E}\left[\sigma_{T}^{2}\vert S_{T}=K\right]}
$$

#### There are two approaches to modelling this relationship

**Parametric**: A parametric representation of the deterministic local
volatility dynamic, A, is specified

**Non-Parametric**: The local volatility (Dupire) is solved explicitly

By solving for the local volatility, it allows the model to re-price
the vanillas at the given maturity

One can estimate $A(t,K)$ with a high degree of precision using 

$A^{2}(t,K)=\frac{\sigma_{D}^{2}(t,K)}{\mathbb{E}\left[\sigma_{t}^{2}\vert S_{t}=K\right]}$
in a forward induction step (for Heston)

$$
\begin{eqnarray*}
\mathbb{E}\left[\sigma_{T}^{2}\vert S_{T}=K\right] & = & \frac{\mathbb{E}\left[\sigma_{T}^{2}\delta(S_{T}-K)\right]}{\delta(S_{T}-K)}\\
 & = & \frac{\int_{0}^{\infty}e^{2y}p(T,\ln K,\,y)dy}{\int_{0}^{\infty}p(T,\ln K,\,y)dy}
\end{eqnarray*}
$$
